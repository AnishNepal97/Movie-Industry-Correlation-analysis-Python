```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r"C:\Users\anish\OneDrive\Desktop\movies.csv")
```

```python
#looking at the data
df.head()
```

```python
#looking for msing data 
#lets loop through the data and see if there is anthing is missing
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#lets explore the data types in our column
print(df.dtypes)
```

```python
#i want to see all rows of data 
pd.set_option('display.max_rows',None)
df.head()
```

```python
df= df.sort_values(by=['gross'],inplace = False ,ascending = False)
df.head()
```

```python
#drop duplicate values if there is any
df.drop_duplicates()
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#lets draw scatterplot with budget vs gross(matplotlib.plt)


plt.scatter(x=df['budget'],y= df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('gross')
plt.ylabel('budget')

```

```python
#lets draw scatterplot with budget vs gross using seaborn 
#regressionplot

sns.regplot(x="gross", y="budget", data=df)
```

```python
sns.regplot(x="score", y="gross", data=df)
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
#df.corr(method ='spearman')
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
#company and star has high corelation for the gross
#now lets looks at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
